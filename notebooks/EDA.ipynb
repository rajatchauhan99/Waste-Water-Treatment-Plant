Let us see count of outliers for each column

To count the number of outliers in each column, we'll define outliers using the interquartile range (IQR) method. Outliers are typically defined as values that fall below

Q1−1.5×IQR or above Q3+1.5×IQR, where Q1 and Q3 are the first and third quartiles, respectively.

```python
import pandas as pd

# Function to count outliers in a column
def count_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return ((column < lower_bound) | (column > upper_bound)).sum()

# Apply the function to each column and store the results
outlier_counts = df.apply(count_outliers)

# Display the count of outliers for each column
outlier_counts
```

To see the lower whisker, upper whisker, and all outlier values for each column, you can calculate these values using the interquartile range (IQR) method. Here's how you can do it:

```python
import pandas as pd

# Function to calculate whiskers and segregate outliers
def whiskers_and_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_whisker = Q1 - 1.5 * IQR
    upper_whisker = Q3 + 1.5 * IQR
    lower_outliers = column[column < lower_whisker]
    upper_outliers = column[column > upper_whisker]
    return lower_whisker, upper_whisker, lower_outliers, upper_outliers

# Apply the function to each column and store the results
results = {}
for col in df.columns:
    lower_whisker, upper_whisker, lower_outliers, upper_outliers = whiskers_and_outliers(df[col].dropna())
    results[col] = {
        "Lower Whisker": lower_whisker,
        "Upper Whisker": upper_whisker,
        "Lower Outliers": lower_outliers.tolist(),
        "Upper Outliers": upper_outliers.tolist(),
        "Count of Lower Outliers": len(lower_outliers),
        "Count of Upper Outliers": len(upper_outliers)
    }

# Display the whiskers and segregated outliers for each column
for col, result in results.items():
    print(f"{col}:")
    print(f"  Lower Whisker: {result['Lower Whisker']}")
    print(f"  Upper Whisker: {result['Upper Whisker']}")
    print(f"  Count of Lower Outliers: {result['Count of Lower Outliers']}")
    print(f"  Lower Outliers: {result['Lower Outliers']}")
    print(f"  Count of Upper Outliers: {result['Count of Upper Outliers']}")
    print(f"  Upper Outliers: {result['Upper Outliers']}")
    print()
```

Removing all these outlers

```python
import pandas as pd

# Function to calculate whiskers
def calculate_whiskers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_whisker = Q1 - 1.5 * IQR
    upper_whisker = Q3 + 1.5 * IQR
    return lower_whisker, upper_whisker

# Remove outliers from each column
for col in df.columns:
    lower_whisker, upper_whisker = calculate_whiskers(df[col].dropna())
    df = df[(df[col] >= lower_whisker) & (df[col] <= upper_whisker)]

# Display the cleaned DataFrame
df
```
