```python
# Assuming y_train is a Pandas Series
scaler_y = MinMaxScaler()

# Convert y_train to a NumPy array and then reshape
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1))

# Similarly, transform y_test
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1))
y_test_scaled
```

1. Linear Regression
Overview: Linear Regression is one of the simplest and most interpretable regression models. It assumes a linear relationship between the input features and the target variable.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the model
model = LinearRegression()

# Train the model
model.fit(X_train_scaled, y_train_scaled)

# Make predictions
y_pred = model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test_scaled, y_pred)
r2 = r2_score(y_test_scaled, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

2. Random Forest Regressor
Overview: Random Forest is an ensemble learning method that builds multiple decision trees and merges them together to get a more accurate and stable prediction.

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train_scaled, y_train_scaled.ravel())  # ravel() is used to flatten the target array

# Make predictions
y_pred = model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test_scaled, y_pred)
r2 = r2_score(y_test_scaled, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

3. Support Vector Regressor (SVR)
Overview: SVR uses the principles of Support Vector Machines but for regression tasks. It tries to fit the best line within a specified margin, ignoring errors as long as they are within that margin.

```python
from sklearn.svm import SVR

# Initialize the model
model = SVR(kernel='rbf')  # 'rbf' kernel is commonly used for regression

# Train the model
model.fit(X_train_scaled, y_train_scaled.ravel())

# Make predictions
y_pred = model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test_scaled, y_pred)
r2 = r2_score(y_test_scaled, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

4. Gradient Boosting Regressor
Overview: Gradient Boosting is another ensemble technique that builds models sequentially, each one correcting the errors of its predecessor. It often yields better performance than Random Forest.

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the model
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)

# Train the model
model.fit(X_train_scaled, y_train_scaled.ravel())

# Make predictions
y_pred = model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test_scaled, y_pred)
r2 = r2_score(y_test_scaled, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

5. Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the model
decision_tree_model = DecisionTreeRegressor(random_state=42)

# Train the model
decision_tree_model.fit(X_train_scaled, y_train_scaled.ravel())

# Make predictions
y_pred_dt = decision_tree_model.predict(X_test_scaled)

# Evaluate the model
mse_dt = mean_squared_error(y_test_scaled, y_pred_dt)
r2_dt = r2_score(y_test_scaled, y_pred_dt)

print(f"Decision Tree Mean Squared Error: {mse_dt}")
print(f"Decision Tree R^2 Score: {r2_dt}")
```

6. Baseline Model or Mean Baseline Model. This model simply predicts the mean value of the target variable for all inputs. It serves as a benchmark to compare more complex models.

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# Calculate the mean of the training target values
y_mean = np.mean(y_train_scaled)

# Predict the mean for all test samples
y_pred_baseline = np.full_like(y_test_scaled, y_mean)

# Evaluate the baseline model
mse_baseline = mean_squared_error(y_test_scaled, y_pred_baseline)
r2_baseline = r2_score(y_test_scaled, y_pred_baseline)

print(f"Baseline Mean Squared Error: {mse_baseline}")
print(f"Baseline R^2 Score: {r2_baseline}")
```
